```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ev=pd.read_csv('data/IEA Global EV Data 2024.csv')
ev
```

```python
ev['year']=pd.to_datetime(ev['year'], format='%Y')
```

```python
# Information about the data frame
print("\nData Frame Info:")
print(ev.info())
```

```python
print(ev.describe(include='all'))
```

```python
# Check for missing values
print(ev.isnull().sum())
```

```python
# Unique values in each column
print("\nUnique Values:")
for column in ev.columns:
    print(f"{column}: {ev[column].nunique()} unique values")
```

```python
# Value counts for categorical columns
print("\nValue Counts for Categorical Columns:")
categorical_columns = ev.select_dtypes(include=['object']).columns
for column in categorical_columns:
    print(f"\n{column} value counts:")
    print(ev[column].value_counts())
```

```python
xls = pd.ExcelFile('data/Hydrogen_production_projects_corrected.xlsx')
df1 = pd.read_excel(xls, 'Projects',skiprows=[0,1,2,4,5,6,7,8,9,10])
df1['Date online']=pd.to_datetime(df1['Date online'], format='%Y')
df1['Decomission date']=pd.to_datetime(df1['Decomission date'], format='%Y')
```

```python
mobility=df1.loc[df1['EndUse_Mobility'].notnull()]
```

```python
mobility
```

```python
print("\nData Frame Info:")
print(mobility.info())
```

```python
print("\nData Description:")
mobility.describe(include='all')
```

```python
mobility.isnull().sum()
```

```python
# Unique values in each column
print("\nUnique Values:")
for column in mobility.columns:
    print(f"{column}: {mobility[column].nunique()} unique values")
```

```python
# Value counts for categorical columns
print("\nValue Counts for Categorical Columns:")
categorical_columns = mobility.select_dtypes(include=['object']).columns
for column in categorical_columns:
    print(f"\n{column} value counts:")
    print(mobility[column].value_counts())
```
